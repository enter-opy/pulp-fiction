```python
import wave
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import IPython
```

```python
def open_wave_file(filename):
    # Open the wave file
    with wave.open(filename, 'rb') as wave_file:
        num_channels = wave_file.getnchannels()
        sample_width = wave_file.getsampwidth()
        sample_rate = wave_file.getframerate()
        num_frames = wave_file.getnframes()


        audio_data = wave_file.readframes(num_frames)
        audio_array = np.frombuffer(audio_data, dtype=np.int16)
        normalized_audio = audio_array / float(2 ** (8 * sample_width - 1))
        normalized_audio = normalized_audio.reshape(-1, num_channels)
        audio_list = normalized_audio.tolist()

        return audio_list, sample_rate
```

```python
filename = 'water-splash.wav'
audio_list, sample_rate = open_wave_file(filename)
print(audio_list)
```

```python
buffer_code = '{{'

for sample in audio_list:
    #print(str(sample[0]))
    buffer_code += str(sample[0])
    buffer_code += ', '

#buffer_code = buffer_code.join('}, {')

for sample in audio_list:
    buffer_code += str(sample[1])
    buffer_code += ', '

buffer_code += '}'

print(buffer_code)
```
